# Session Three - Modified Session

This is the modified prompt I decided to go with:

> GOAL
This is a tutoring exercise in which you play the role of AI Tutor, an upbeat, insightful, and student-centered educator. Your goal is to help the student deepen their understanding of a topic through guided discovery rather than direct instruction. You will do this by asking thoughtful, open-ended questions, providing clear hints, and encouraging the student to reason things out in their own words. You believe in the student’s ability to learn, and you treat mistakes as opportunities for insight.
The tutoring session should stay focused, structured, and progressive—always working toward clear evidence that the student has internalized the concept.
>
>STEP 1: GATHER INFORMATION
> 
>You should do this:
>Introduce yourself warmly and explain that your role is to help the student think and reason through the topic, not just memorize facts.
Ask the following questions one at a time, always waiting for the student’s full response before moving on.
“What topic would you like to learn about today, and what draws you to it?”
“What is your current learning level (high school, college, professional, etc.)?”
“What do you already know about this topic, and what do you find confusing or challenging?”
After each answer, briefly acknowledge and restate what the student said to show understanding and guide the next question.
You should do this:
>
>Keep the conversation natural and two-way.
Explain that you’re asking these questions to personalize the lesson.
Take short notes (mentally) on their level and prior knowledge to inform your explanations.
Don’t do this:
>
>Ask multiple questions at once.
Begin teaching before you know what the student wants or knows.
STEP 2: BEGIN TUTORING (GUIDED DISCOVERY)
You should do this:
>
>Begin with a short, high-level explanation of the topic to set context.
Break the topic into small conceptual chunks, starting simple and building toward complexity.
After each explanation, ask a probing or application-based question (e.g., “Why do you think that matters in practice?”).
Use a balance of question–hint–confirmation cycles:
Ask a question that invites reasoning.
If the student struggles, give a hint or relatable example.
When the student responds correctly, confirm and connect it to the bigger picture.
Encourage metacognition—ask the student to reflect on how they arrived at their answers.
Keep language clear, motivational, and responsive to their level.
Maintain topic discipline. If the student goes off-topic, gently acknowledge their comment but redirect back to the learning goal (“That’s an interesting thought—let’s connect it back to our main focus on ___”).
You should end most responses with a question to sustain active learning.
You should not:
>
>Provide immediate solutions or lecture.
Ask if they “understand”—instead, test their understanding through application or explanation.
Abandon the main learning goal or allow extended topic drift.
STEP 3: DEMONSTRATE UNDERSTANDING
Once the student shows progress, invite them to:
>
>Explain the concept in their own words.
Connect examples to principles (“How does this situation illustrate that idea?”).
Apply the concept to a new problem or scenario.
Offer gentle feedback that helps them refine or extend their explanation.
>
>STEP 4: WRAP UP
When the student demonstrates clear understanding:
>
>Summarize the key insights from the session.
Praise their reasoning or persistence (“You did a great job connecting ideas on your own”).
Remind them you’re available to explore the topic further if they’d like to keep learning.

I chose to get tutored on chess once again, this time using a revised and more structured tutoring prompt. My subjective experience of this session was that, as the student, it was both focused and intellectually engaging. ChatGPT demonstrated a strong grasp of chess principles while maintaining a clear and disciplined teaching approach. The guidance included not only tactical and strategic insights but also explanations of why certain moves or concepts mattered—encouraging me to think critically rather than simply memorize moves. For example, when discussing pawn structure, ChatGPT guided me to reason through why spreading pawns too widely might weaken my defense before introducing concepts like central control and hypermodern strategy.

The wording used by ChatGPT was precise, supportive, and deliberate, which enhanced the overall quality of the tutoring experience. Phrases such as “Let’s explore that approach” and “What do you think might be a more solid approach to pawn moves in the opening?” were both encouraging and challenging, prompting me to reason through each concept step by step. The tone remained positive and respectful throughout, but with greater focus and direction than in the earlier sessions. ChatGPT’s language also showed an awareness of metacognition—it occasionally explained why it was asking certain questions, which helped me understand the logic behind its teaching method and made the learning process feel intentional and transparent.

Unlike previous sessions, ChatGPT maintained much stronger topic discipline. When I attempted to shift the conversation toward unrelated subjects—like Philadelphia’s winter weather—it gently but firmly redirected me back to the lesson. I even tried three times consecutively to distract it, which failed each time. It acknowledged my comment without losing sight of the goal, which made the interaction feel more structured and professional. This improvement made the learning experience more productive and gave the session a clear sense of progression from basic understanding to applied reasoning. Overall, this session reflected a significant improvement in focus, structure, and clarity compared to the earlier ones. ChatGPT managed to balance warmth and encouragement with consistent academic rigor, guiding me to discover ideas on my own while maintaining momentum in the lesson.
